```python
import requests
from datetime import datetime, timedelta
import json

# Function to generate dates for the 6-month period
def generate_dates(start_date, end_date):
    for n in range(int((end_date - start_date).days) + 1):
        yield start_date + timedelta(n)

# Define the base URL
base_url = "https://api.weatherstack.com/historical"

# End date is set to November 1, 2023, and start date is 6 months before
end_date = datetime.strptime("2023-11-01", "%Y-%m-%d")
start_date = end_date - timedelta(days=2*365)  # Approximately 5 years

# Initialize an empty list to store all daily data
all_data = []

# Iterate over each day in the 6-month period
for single_date in generate_dates(start_date, end_date):
    # Format the date as required by the API
    formatted_date = single_date.strftime("%Y-%m-%d")

    # Define parameters for the API request
    params = {
        "access_key": "ba08048a4126e6ff124444ffc539b9e2",
        "query": "Washington",
        "historical_date": formatted_date,
        "hourly": "1"
    }

    # Make the GET request
    response = requests.get(base_url, params=params)

    # Check if the request was successful
    if response.status_code == 200:
        # Append the JSON response to the all_data list
        all_data.append(response.json())
    else:
        print("Error on", formatted_date, ":", response.status_code, response.text)

# Convert the list of all data to a JSON string
json_string = json.dumps(all_data, indent=4)  # indent parameter for pretty-printing

# Specify the file path where you want to save the JSON data
file_path = "weather-output.json"

# Write the JSON string to the file
with open(file_path, "w") as file:
    file.write(json_string)

print(f"JSON data saved to {file_path}")
```

```python
file_path = 'weather-output.json'

# Read the JSON file into a DataFrame
weather_df = pd.read_json(file_path)

# Display the first few rows of the DataFrame
weather_df.head()
```

```python
# Extract the 'historical' column which contains the data we are interested in
historical_data = [day[next(iter(day))] for day in weather_df['historical']]

# Convert the list of dictionaries to a DataFrame
historical_df = pd.json_normalize(historical_data)

# Display the first few rows of the new DataFrame
historical_df
```

```python
import pandas as pd

# Initialize an empty DataFrame to collect the daily data
daily_data = []

# Loop through each row in the historical data to collect the daily summaries
for index, row in historical_df.iterrows():
    # Get the date from the row
    date = row['date']
    
    # Create a dictionary to hold the daily data
    daily_summary = {'date': date}
    
    # Initialize a dictionary to keep the sum of numeric values for each key
    numeric_sums = {}
    # Initialize a counter for the number of hourly data points to calculate the average
    count = 0

    # Since each 'hourly' entry is a list of dictionaries, we'll iterate through and collect the sum of numeric values
    for key in row['hourly'][0].keys():
        # Skip the 'time' key since we are aggregating daily data
        if key != 'time':
            # Check if the value is numeric and sum it up
            if isinstance(row['hourly'][0][key], (int, float)):
                numeric_sums[key] = sum(hour.get(key, 0) for hour in row['hourly'])
                count = len(row['hourly'])
    
    # Calculate the average for each numeric attribute
    for key, total in numeric_sums.items():
        daily_summary[key] = total / count
    
    # Append the daily summary to the list
    daily_data.append(daily_summary)

# Convert the list of daily summaries to a DataFrame
daily_df = pd.DataFrame(daily_data)

# Set the date as the index of the DataFrame
daily_df.set_index('date', inplace=True)

# Display the first few rows of the daily DataFrame
daily_df.head()
```

```python
# Since the 'hourly' column contains lists of dictionaries, we'll need to flatten these
# We will create a new DataFrame for the hourly data
hourly_data = []

# Loop through each row in historical data and collect the hourly data
for index, row in historical_df.iterrows():
    # Each 'hourly' entry is a list of dictionaries
    for hour in row['hourly']:
        # Add the date to the hourly data
        hour['date'] = row['date']
        hourly_data.append(hour)

# Create a DataFrame with the hourly data
hourly_df = pd.DataFrame(hourly_data)

# Since the 'date' and 'time' columns together represent a unique timestamp, we'll set them as a multi-index
hourly_df.set_index(['date', 'time'], inplace=True)

# Display the first few rows of the hourly DataFrame
hourly_df.head()
```

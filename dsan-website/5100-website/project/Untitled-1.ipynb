```python
import requests
import json
import pandas as pd

url = "https://apidojo-forever21-v1.p.rapidapi.com/categories/v2/list"

headers = {
	"X-RapidAPI-Key": "14d8d40c8amsh63b4f7a4468db90p1c3237jsn9e3b7c33366d",
	"X-RapidAPI-Host": "apidojo-forever21-v1.p.rapidapi.com"
}

response = requests.get(url, headers=headers)

print(response.json())
```

```python
data = response.json()
print(data)

#Convert the data to a JSON string
json_string = json.dumps(data, indent=4)  # indent parameter for pretty-printing

# Specify the file path where you want to save the JSON data
file_path = "forever-21_data.json"

# Write the JSON string to the file
with open(file_path, "w") as file:
    file.write(json_string)

print(f"JSON data saved to {file_path}")
```

```python
with open('forever-21_data.json', 'r') as f:
  data = json.load(f)

str_json = json.dumps(data)

# Getting dictionary
json_dict = json.loads(str_json)

# Pretty Printing JSON string back
data_output = json.dumps(json_dict, indent = 4, sort_keys=True)
print(data_output)
```
